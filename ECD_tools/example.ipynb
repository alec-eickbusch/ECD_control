```python
import qutip as qt
import numpy as np
import matplotlib.pyplot as plt
from ECD_control.ECD_tools.ECD_tools import *
from bosonic_codes.GKPCode import SquareGKPCode
```

```python
N = 40
N2 = 2
a = qt.tensor(qt.identity(N2),qt.destroy(N))
q = qt.tensor(qt.destroy(N2),qt.identity(N))
```

```python
alpha = 2 + 1j
c = qt.displace(N,alpha)*qt.basis(N,0)
psi = qt.tensor(qt.basis(N2,1),c)
theta = (np.pi/2.0)
R = (1j*theta*a.dag()*a).expm()
psi = R*psi
```

```python
xvec = np.linspace(-4,4,101)
plot_wigner(psi.ptrace(1), xvec=xvec)
```

```python
N = 100
Delta = 0.35
c = SquareGKPCode(Delta, N)
z = c.plusz()
y = c.plusy()
```
